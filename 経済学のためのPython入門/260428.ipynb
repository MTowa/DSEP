zip関数

```python
c_lst = []
y_lst = [1, 2, 3, 4, 5]
mpc_lst = [0.6, 0.5, 0.5, 0.3, 0.2]

for y, mpc in zip(y_lst, mpc_lst):
    c = 0.1 + mpc * y
    c_lst.append(c)

c_lst
```

資産の推移

```python
wealth = float(input("資産を入力してね"))
wealth_lst = [wealth]

for _ in range(5):
    wealth *= (1 + 0.03)
    wealth_lst.append(wealth)

wealth_lst
```

壁までの距離

```python
distance = float(input("距離を入力してね"))
count = 0

while distance > 0.001:
    print(distance)
    distance -= distance/2
    count += 1

print(f"ループの回数：{count}")
print(f"残りの距離：{distance*1000}mm")
```
