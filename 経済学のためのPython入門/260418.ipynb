**将来価値**

```python
def future_value(x, r=0.03, t=10):
    result = x * (1+r) ** t
    return result

shikin = input("資金を入力: ")
shikin = int(shikin)
future_value(shikin)
```

**需要・供給モデル**

```python
def demand(p):
    return 40 / (p ** (0.1))

def supply(p):
    return p ** (1.5)

def d_larger_than_s(p):

    d = demand(p)
    s = supply(p)

    print(f"価格={p}")
    print(f"需要={d:.2}")
    print(f"供給={s:.2}")

    return d > s


price = int(input("価格を入力："))
d_larger_than_s(price)

```
